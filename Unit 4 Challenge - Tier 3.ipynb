```python
# We want to have a row of observations for every borough. From the previous call to properties.head() we see that
# boroughs are sorted in columns. We proceed to perform a transposition on the dataset 'properties' and check that 
# the names of boroughs now label the rows:

properties = properties.transpose()
properties.head()
```

```python
# We would like now to have the dates as the columns names. In order to do this we first reassign the dates to the
# columns names:
properties.columns = properties.iloc[0]

# Now we eliminate the row at index 0, which is redundant:
properties = properties.drop(0)

# Let's see what our dataframe looks like now:
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# We want to rename the first two columns. "Unnamed: 0" should be renamed as "London Borough" and "NaN" as "ID":
properties = properties.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})

properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# We want to transform our dataframe in such a way that we only have one column representing the average
# house prices within a given month for a given year. For this purpose we use .melt() method:

properties = pd.melt(properties, id_vars= ['London_Borough', 'ID'])
properties.head()

# And now rename the last two columns as "Month" and "Average_price"
properties = properties.rename(columns = {0: 'Month', 'value': 'Average_price'})
properties.head()

# Notice that the list of boroughs will repeat itself as we scroll down the dataframe until every month contained
# in the original dataset has been displayed.
```

```python
#The Average_price should be numeric. Let's convert it to float type:
properties['Average_price'] = pd.to_numeric(properties['Average_price'])

# Recheck columns data types
properties.dtypes
```

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# The do not contain any information so we will delete them. We shall use dropna.() method on our dataset:

properties = properties.dropna()
properties.head(50)
```

```python
# Let's now check that every column has the same number of elements:

properties.count()
```

```python
# Now we want to get rid of the rows that do not represent proper boroughs. Here is the list of values that
# do not correspond to boroughs:

non_Boroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
# Let's eliminate the rows containing these values by filtering:

properties = properties[~properties.London_Borough.isin(non_Boroughs)]
properties.head(50)
```

```python
# Our dataframe is ready to be operated with. We shall rename it as 'df':

df = properties
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# We shall focus on the Lambeth borough. Let's define the variable 'line_plot' whichi plots the average prices on 
# the y-axis versus the month on the x-axis:

line_plot = df[df['London_Borough'] == 'Lambeth'].plot(kind ='line', x = 'Month', y='Average_price')

# Set the label of the y-axis. 
line_plot.set_ylabel('Price')
```

```python
# Now check average prices for 1998
print(df_avg[df_avg['Year']==1998])

# and for 2018
print(df_avg[df_avg['Year']==2018])
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Define the create_price_ratio function

def create_price_ratio(d):
    avg_price_1998 = float(d['Average_price'][d['Year']==1998])
    avg_price_2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [avg_price_2018/avg_price_1998]
    return ratio

# Create a dictionary that will store the ratios as values and boroughs as keys

ratios = {}

# We call create_price_ratio and compute the ratio for each unique value of the column "London_Borough" in our dataframe

for b in df_avg['London_Borough'].unique():
    
    borough = df_avg[df_avg['London_Borough'] == b]
    
    ratios[b] = create_price_ratio(borough)
    
# Print out the dictionary "ratios"
print(ratios) 

# Convert the dictionary "ratios" into the dataframe "df_ratios":

df_ratios = pd.DataFrame(ratios)

# Let's have a look at it:

df_ratios.head()
```

```python
# Give format to df_ratios. First we transpose:

df_ratios = df_ratios.transpose()
df_ratios.head()

# Now we reset index

df_ratios = df_ratios.reset_index()
df_ratios.head()

# Finally we rename the "Index" column as "London_Borough" and the "0" column as "ratio_2018"

df_ratios.rename(columns={'index':'London_Borough', 0:'ratio_2018'}, inplace=True)
df_ratios.head()
```

```python
# Now we sort the ratios in descending order:

sorted_ratios = df_ratios.sort_values(by='ratio_2018',ascending=False)

sorted_ratios.head(15)
```

```python
# Make a bar plot with the ratios on the y-axis and the boroughs on the x-axis:

bar_plot = sorted_ratios[['London_Borough','ratio_2018']].plot(kind='bar')

# Label the x-axis as the names of the boroughs
bar_plot.set_xticklabels(sorted_ratios.London_Borough)
```
